```python
import matplotlib.pyplot as plt
import numpy as np

data = {
    "Feature": ["PropertyValue", "ShortwaveRadiation", "Precipitation", "Evapotranspiration", "AQI", "Temperature",
                "Population", "NightIndustryIncome", "IndustryEmploy", "AverageCommuteTime", "IndividualCount"],
    "A": [0.174869, 0.017518, 0.014994, 0.011641, 0.028644, 0.012079, 0.266229, 0.151787, 0.171379, 0.122693,
          0.028168],
    "B": [0.112802, 0.074389, 0.112536, 0.095424, 0.075938, 0.163128, 0.135849, 0.034339, 0.041617, 0.043868,
          0.11011],
    "C": [0.106981, 0.004578, 0.006192, 0.004796, 0.034916, 0.027958, 0.16964, 0.202913, 0.285821, 0.150373,
          0.005832],
    "Group Decision": [0.131550667, 0.032161667, 0.044574, 0.037287, 0.046499333, 0.067721667, 0.190572667, 0.129679667,
                       0.166272333, 0.105644667, 0.048036667]
}

# Set the color map and colors for each stack
cmap = plt.get_cmap('tab20c')
colors = [cmap(i) for i in range(4)]

# Set up the figure and subplot
fig, ax = plt.subplots(figsize=(10, 6))

# Set the position and width of the bars
barWidth = 0.75
bars = [0] * len(data["Feature"])

# Plot the A stack
for i, val in enumerate(data["A"]):
    bars[i] = ax.bar(data["Feature"][i], val, bottom=bars[i], width=barWidth, color=colors[0])

# Plot the B stack
for i, val in enumerate(data["B"]):
    bars[i] = ax.bar(data["Feature"][i], val, bottom=bars[i], width=barWidth, color=colors[1])

# Plot the C stack
for i, val in enumerate(data["C"]):
    bars[i] = ax.bar(data["Feature"][i], val, bottom=bars[i], width=barWidth, color=colors[2])

# Plot the Group Decision stack
for i, val in enumerate(data["Group Decision"]):
    bars[i] = ax.bar(data["Feature"][i], val, bottom=bars[i], width=barWidth, color=colors[3])

# Add legend and axis labels
ax.legend(["A", "B", "C", "Group Decision"])
ax.set_ylabel("Proportion")
ax.set_xlabel("Features")

# Show the plot
plt.show()
```
